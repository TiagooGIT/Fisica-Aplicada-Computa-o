```python

y0 = [0]


t = np.linspace(0,100)

b2 = 1
f2 = 4
y2 = 2
n1 = 5
y1 = 4
g1 = 2
k1 = 3
g2 = 4
k2 = 2
n2 = 4
Y1 = y1/n1
Y2 = y2/n2
n = n2 / n1
m1 = k1
m2 = k2
f1 = 2
y3 = n1
y4 = n2




parametros = [b2,f2,y2,n1,y1,g1,k1,g2,k2,n2, Y1, Y2, n, m1, m2,f1,y3,y4]
```

```python


def number_of_infectous(pop,t, parametros):
    
    p = pop[0]
    
    
    
    b2 = parametros[0]
    f2 = parametros[1]
    y2 = parametros[2]
    n1 = parametros[3]
    y1 = parametros[4]
    
    
    dydt = ((b2*f2*y2*(n1 - y1)) * t) / n1

    return dydt


def infectous_Humans(pop1, t, parametros):    
    
    p1 = pop1[0]
    
    
    
    b2 = parametros[0]
    f2 = parametros[1]
    y2 = parametros[2]
    n1 = parametros[3]
    y1 = parametros[4]
    g1 = parametros[5]
    k1 = parametros[6]
    
    dy1dt = (((b2*f2*y2*(n1 - y1)) / n1) - (g1 + k1)*y1 )     # No todos quantos estao infetados, consoante o numero de recuperados
    
    
    return dy1dt



def infectous_Mosquitos(pop2, t, parametros): 
    
    p2 = pop2[0]
    
    
    
    b2 = parametros[0]
    f2 = parametros[1]
    y2 = parametros[2]
    n1 = parametros[3]
    
    g2 = parametros[4]
    k2 = parametros[5]
    n2 = parametros[6]
    
    dy2dt = (((b2*f2*y2*(n2 - y2)) / n1) - (g2 + k2)*y2 )    # No todos quantos estao infetados, consoante o numero de recuperados
    
    
    return dy2dt


def newBorn_Humans(pop3, t, parametros): 
    
    p2 = pop3[0]
    
    
    
    b2 = parametros[0]
    f2 = parametros[1]  
    Y1 = parametros[2]
    Y2 = parametros[3]
    m1 = parametros[4]
    
    dY1dt = ((b2*f2*Y2*n*(1 - Y1)) - (m1 * Y1))     # No todos quantos estao infetados, consoante o numero de recuperados
    
    
    return dY1dt

def newBorn_Mosquitos(pop4, t, parametros): 
    
    p4 = pop4[0]
    
    
    
    b2 = parametros[0]
    f1 = parametros[1]
    Y1 = parametros[2]
    f1 = parametros[3]
    Y2 = parametros[4]
    m2 = parametros[5]
    
    
    dY2dt = ((b2*f1*Y1*(1 - Y2)) - (m2 * Y2))     # No todos quantos estao infetados, consoante o numero de recuperados
    
    
    return dY2dt


```

```python

def Equilibria(pop5, t, parametros): 
    
    p5 = pop5[0]
    
    
    
    b2 = parametros[0]
    n = parametros[1]
    f1 = parametros[2]
    f2 = parametros[3]
    g1 = parametros[4]
    m2 = parametros[5]
    
    
    P = ((n*b2^2*f1*f2)/ (g1*m2))     # No todos quantos estao infetados, consoante o numero de recuperados
    
    
    
    return P

## Equilibrio Trivial
##Todas as equaçoes forma divididas por Y1*Y2 como referido no enunciado
## Todos os valores obtidos sao referentes aos valores atribuidos ás variaveis + acima

T1 = infectous_Humans(y0, t, parametros)/Y1*Y2
T2 = infectous_Mosquitos(y0, t, parametros)/Y1*Y2
T3 = newBorn_Humans(y0, t, parametros)/Y1*Y2
T4 = newBorn_Mosquitos(y0, t, parametros)/Y1*Y2
#T5 = Equilibria(y0, t, parametros)/Y1*Y2
print("Verificação Equilibrio se < 1, humanos infetados:",T3)
print ("Verificação Equilibrio se < 1, Mosquitos infetados:",T4)
print ("Verificação Equilibrio se < 1, Nascimentos(Humanos):",T2)
print ("Verificação Equilibrio se < 1, Nascimentos (Mosquitos):",T1)

plt.axhline(T4,marker='x',linestyle= '-', linewidth=0.8, color='black', label='Number of Infectous - Dt')
plt.axhline(T3,marker='x',linestyle= '-', linewidth=0.8, color='black', label='Number of Infectous - Dt')
plt.axhline(T2,marker='x',linestyle= '-', linewidth=0.8, color='black', label='Number of Infectous - Dt')
plt.axhline(T1,marker='x',linestyle= '-', linewidth=0.8, color='black', label='Number of Infectous - Dt')
```
